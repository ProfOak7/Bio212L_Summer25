```python
import pandas as pd
import statistics
from statistics import mean, stdev
from scipy.stats import sem
import scipy.stats as stats
import matplotlib.pyplot as plt
#include if using a jupyter notebook, remove if using a .py file
%matplotlib inline
```

```python
#Optional for disabling warnings
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
# Load the dataset
df = pd.read_excel(r'./Bio212L_Lab1.xlsx', index_col=0)
df
```

```python
#Assign the names to the following columns:
Before = df.Before
Slow = df.Slow
Fast = df.Fast
```

```python
#Calculate the mean and standard error of mean (SEM)
BEFORE_MEAN = mean(Before)
BEFORE_SEM = sem(Before)
SLOW_MEAN = mean(Slow)
SLOW_SEM = sem(Slow)
FAST_MEAN = mean(Fast)
FAST_SEM = sem(Fast)
```

```python
# build a bar plot
plt.bar(['Before Exercise', 'Slow Exercise (30 sec)', 'Fast Exercise (30 sec)' ],[BEFORE_MEAN, SLOW_MEAN, FAST_MEAN], yerr=[BEFORE_SEM, SLOW_SEM, FAST_SEM], capsize=10)
plt.ylabel('HR (BPM)')
plt.title('Effect of Exercise on Heart Rate')
```
